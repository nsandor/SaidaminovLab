```python
#filename = input('Enter filename\n')
filename = "/Users/yukiharuta/Desktop/Temp-Data/MAPbBr3-Paper/230703-ANALYSIS-First90min.xlsx"
ans="Evap (g/h)"

import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns #

df = pd.read_excel(filename, index_col=0)

sns.pairplot(df) #
plt.show()
```

```python
from sklearn.ensemble import RandomForestRegressor #
from sklearn.model_selection import train_test_split #
import numpy as np
from sklearn.metrics import r2_score            # 
from sklearn.metrics import mean_squared_error

col=df.columns.drop([ans])
# col=df.columns.drop('ID')
X, Y = df[col], df[ans]

test_sizes = [0.1,0.3,0.5]
Errors = []
for test_size in test_sizes:
    for i in range(100):
        X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=test_size, random_state=1) #X,YAI
        RF= RandomForestRegressor(n_estimators=200)
        RF.fit(X_train, Y_train)
        RF.fit(X, Y)

        Y_test_pred  = RF.predict(X_test)
        Y_train_pred = RF.predict(X_train)
        
        Error = [a-b for a,b in zip(Y_test_pred, Y_test)]
        for e in Error:
            Errors.append(e)

    # (RMSE)
    print('RMSE train: %.2f, test: %.2f' % (
            mean_squared_error(Y_train, Y_train_pred, squared=False), # 
            mean_squared_error(Y_test, Y_test_pred, squared=False)    # 
          ))

    # (R^2)
    print('R^2 train: %.2f, test: %.2f' % (
            r2_score(Y_train, Y_train_pred), # 
            r2_score(Y_test, Y_test_pred)    # 
          ))
    plt.figure(figsize=(6,6))
    plt.hist(Errors, label = str(test_size), density=True)
    plt.legend()
    plt.show()
#     plt.figure(figsize=(6,6))
#     plt.scatter(Y_test,Y_test_pred)
#     plt.xlabel(ans)
#     plt.ylabel(ans+" predicted")
#     plt.show()

#     plt.scatter(Y_test,Error)
#     plt.xlabel(ans)
#     plt.ylabel('Error')
#     plt.show()
```
